```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from sklearn.metrics.pairwise import cosine_distances

sw_indo = stopwords.words('indonesian') + list(punctuation)
```

```python
df = pd.read_csv('berita_terkini.csv')
df.head()
```

```python
def remove_punctuation(text):
    trans_table = str.maketrans('', '', punctuation)
    return text.translate(trans_table).lower()

text = df['Judul Berita'].apply(remove_punctuation)
text
```

```python
encoder = CountVectorizer(stop_words=sw_indo, tokenizer=word_tokenize)
bank_title = encoder.fit_transform(df['Judul Berita'])
bank_theme = encoder.transform(df['Theme'])
```

```python
idx = 0
content = df.loc[idx, 'Judul Berita']
code = encoder.transform([content])
dist = cosine_distances(code, bank_title)
rec_idx = dist.argsort()[0, 1:(10+1)]
```

```python
df.loc[rec_idx]
```
